### Task 1




##### Display the names of the unique launch sites  in the space mission


```python
query = "SELECT DISTINCT Launch_Site FROM SPACEXTBL"

# Execute the query and fetch the results
launch_sites = pd.read_sql_query(query, con)
```


### Task 2


#####  Display 5 records where launch sites begin with the string 'KSC' 


```python
query="SELECT Launch_Site from SPACEXTBL WHERE Launch_Site LIKE '%KSC%' LIMIT 5"
records = pd.read_sql_query(query, con)
records
```

# Task 3




##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
query="SELECT SUM(PAYLOAD_MASS__KG_) from SPACEXTBL WHERE Customer=='NASA (CRS)'"
total_payload=pd.read_sql_query(query, con)
total_payload
```

### Task 4




##### Display average payload mass carried by booster version F9 v1.1


```python
query="SELECT AVG(PAYLOAD_MASS__KG_) from SPACEXTBL WHERE Booster_Version=='F9 v1.1'"
avg_payload=pd.read_sql_query(query, con)
avg_payload
```

### Task 5

##### List the date where the succesful landing outcome in drone ship was acheived.


_Hint:Use min function_ 


```python
query="SELECT Date from SPACEXTBL WHERE Landing_Outcome=='Success (drone ship)'"
date=pd.read_sql_query(query, con)
date
```

### Task 6

##### List the names of the boosters which have success in ground pad  and have payload mass greater than 4000 but less than 6000


```python
query="SELECT Payload from SPACEXTBL WHERE Landing_Outcome=='Success (ground pad)' AND PAYLOAD_MASS__KG_ BETWEEN 4000 AND 6000"
booster_names=pd.read_sql_query(query, con)
booster_names
```

### Task 7




##### List the total number of successful and failure mission outcomes


```python
query="SELECT Mission_Outcome, COUNT(Mission_Outcome) FROM SPACEXTBL GROUP BY Mission_Outcome"
total=pd.read_sql_query(query, con)
total
```

### Task 8



##### List the   names of the booster_versions which have carried the maximum payload mass. Use a subquery


```python
query="SELECT Payload from SPACEXTBL WHERE PAYLOAD_MASS__KG_ =(SELECT MAX(PAYLOAD_MASS__KG_) FROM SPACEXTBL)"
names=pd.read_sql_query(query,con)
names
```

### Task 9


##### List the records which will display the month names, succesful landing_outcomes in ground pad ,booster versions, launch_site for the months in year 2017
**Note: SQLLite does not support monthnames. So you need to use substr(Date,6,2) for month, substr(Date,9,2) for date, substr(Date,0,5),='2017' for year.**


```python
query='''SELECT 
    CASE substr(Date, 6, 2)
        WHEN '01' THEN 'January'
        WHEN '02' THEN 'February'
        WHEN '03' THEN 'March'
        WHEN '04' THEN 'April'
        WHEN '05' THEN 'May'
        WHEN '06' THEN 'June'
        WHEN '07' THEN 'July'
        WHEN '08' THEN 'August'
        WHEN '09' THEN 'September'
        WHEN '10' THEN 'October'
        WHEN '11' THEN 'November'
        WHEN '12' THEN 'December'
    END AS Month,
    BOOSTER_VERSION,
    LAUNCH_SITE,
    LANDING_OUTCOME
FROM SPACEXTBL
WHERE 
    substr(Date, 0, 5) = '2017' 
    AND Mission_Outcome = 'Success'
    AND Landing_Outcome LIKE '%Ground Pad%'
'''
records=pd.read_sql_query(query,con)
records
```

### Task 10




##### Rank the count of landing outcomes (such as Failure (drone ship) or Success (ground pad)) between the date 2010-06-04 and 2017-03-20, in descending order


```python
query='''SELECT Landing_Outcome, COUNT(Landing_Outcome) AS Outcome_Count
FROM SPACEXTBL
WHERE (Landing_Outcome = 'Failure (drone ship)' OR Landing_Outcome = 'Success (ground pad)')
  AND DATE BETWEEN '2010-06-04' AND '2017-03-20'
GROUP BY Landing_Outcome
ORDER BY Outcome_Count DESC
'''
data=pd.read_sql_query(query,con)
data
```
